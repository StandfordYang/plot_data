```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.animation import ArtistAnimation
from matplotlib import animation
from IPython.display import HTML
```

```python
# %matplotlib notebook
fig, ax = plt.subplots(figsize=(9, 6))
fig.set_facecolor('gray')
allcolor = ['orange', 'blue', 'orange', 'blue']
_ = ax.pie([1, 1, 1, 1], colors=allcolor, startangle=10,  wedgeprops=dict(width=0.3))
_ = ax.pie([1, 1, 1, 1], colors=allcolor, startangle=10-45,  
           wedgeprops=dict(width=0.006), radius=0.7-0.006)
_ = ax.pie([1, 1, 1, 1], colors=allcolor, startangle=10+45,  
           wedgeprops=dict(width=0.007), radius=1.007)

def animate(angle):
    angle = - angle * 10

    _ = ax.pie([1, 1, 1, 1], colors=allcolor, startangle=angle,  wedgeprops=dict(width=0.3))
    _ = ax.pie([1, 1, 1, 1], colors=allcolor, startangle=angle-45,  
           wedgeprops=dict(width=0.006), radius=0.7-0.006)
    _ = ax.pie([1, 1, 1, 1], colors=allcolor, startangle=angle+45,  
           wedgeprops=dict(width=0.007), radius=1.007)
    print(f"angle: {angle}")



ani = FuncAnimation(fig, animate, frames=72, interval=20)
ani
# ani.save(filename='mypie.mp4')
# plt.show()
from IPython.display import HTML
# HTML(ani.to_html5_video())
# ani.save(filename='mypie.mp4', savefig_kwargs=dict(facecolor='gray'))
ani.save('mypie.gif', writer="pillow", savefig_kwargs=dict(facecolor='gray'))
```
